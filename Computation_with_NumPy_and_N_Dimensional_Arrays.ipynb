```python
array_2d = np.array([[1, 2, 3, 9], 
                     [5, 6, 7, 8]])

print(f'Dimensions: {array_2d.ndim}')
print(array_2d.shape)
print(f'array_2d has {array_2d.shape[0]} rows and {array_2d.shape[1]} columns')
array_2d
```

# Linear Algebra with Vectors

```python
v1 = np.array([4, 5, 2, 7])
v2 = np.array([2, 1, 3, 3])
# Add the elements of the 2 vectors. Works because these are Numpy arrays
v1 + v2
```

```python
# Python Lists vs ndarrays
list1 = [4, 5, 2, 7]
list2 = [2, 1, 3, 3]

# Where as in regular Python, the same operation will concatenate the lists.
list1 + list2
```

```python
# Same concept as above, but with multiplication
v1 * v2
```

```python
# Of course this doesn't work with regular lists.
list1 * list2
```

# Broadcasting and Scalars


```python
# Oftentimes you'll want to do some sort of operation between an array and a single number. 
# In mathematics, this single number is often called a scalar. For example, you might want 
# to multiply every value in your NumPy array by 2. to achieve this result, NumPy will make 
# the shape of the smaller array, our scalar, compatible with the larger array. This is what 
# the documentation refers to when it mentions the term "broadcasting". 
array_2d = np.array([[1, 2, 3, 4], 
                     [5, 6, 7, 8]])
```

```python
array_2d + 10
```

```python
array_2d * 5
```

# Matrix Multiplication with @ and .matmul()

<img src=https://i.imgur.com/LCdhmi8.png width=350>

```python
a1 = np.array([[1, 3],
               [0, 1],
               [6, 2],
               [9, 7]])

b1 = np.array([[4, 1, 3],
               [5, 8, 5]])

print(f'{a1.shape}: a has {a1.shape[0]} rows and {a1.shape[1]} columns.')
print(f'{b1.shape}: b has {b1.shape[0]} rows and {b1.shape[1]} columns.')
print('Dimensions of result: (4x2)*(2x3)=(4x3)')
```

```python
# c12 should be (1x1) + (3x8) = 25
# c33 should be (6x3) + (2x5) = 28
# Result of this multiplication should be, 4x2 X 2x3 -> 4x3 matrix
# i.e. mxn X nxp -> mxp
```

**Challenge**: Convert the image to black and white. The values in our `img` range from 0 to 255. 
* Divide all the values by 255 to convert them to sRGB, where all the values are between 0 and 1. 
* Next, multiply the sRGB array by the `grey_vals` to convert the image to grey scale. 
* Finally use Matplotlib's [`.imshow()`](https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.imshow.html) together with the colormap parameter set to gray `cmap=gray` to look at the results. 

```python
# These grey_vals are from a 'Y linear' formula for converting color images to greyscale
# https://en.wikipedia.org/wiki/Grayscale#Colorimetric_(perceptual_luminance-preserving)_conversion_to_grayscale
# These values are sRGB, but our img array is RGB (0-255), so we have to convert the values in our img array.
grey_vals = np.array([0.2126, 0.7152, 0.0722])
sRGB_img = img / 255
sRGB_img
```

```python
# Using the shorthand instead of np.matmult(sRGB_img, grey_vals)
grey_img = sRGB_img @ grey_vals
```

```python
# cmap='grey' is required or the function will not know it's a black and white image.
# Without this parameter, the image will be green.
plt.imshow(grey_img, cmap='grey')
```
